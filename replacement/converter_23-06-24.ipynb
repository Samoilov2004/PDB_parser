```python
from rdkit import Chem
import pandas as pd
import warnings
import os

warnings.filterwarnings("ignore")
```

```python
Names = []
smiles = []
inf = open('Components-pub.sdf','rb')
fsuppl = Chem.ForwardSDMolSupplier(inf)
for mol in fsuppl:
    if mol != None:
        name = mol.GetProp('_Name')
        smi = Chem.MolToSmiles(mol)
        
        Names.append(name)
        smiles.append(smi)

inf.close()
```

```python
data = {'id': Names, 'smiles_original': smiles}
df = pd.DataFrame.from_dict(data)
df
```

```python
df.to_excel("input_pdb.xlsx", index=False)
```
